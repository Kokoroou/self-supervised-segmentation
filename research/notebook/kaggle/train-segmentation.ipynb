```python
import time
import wget


# Download checkpoint directly from kernel to reduce time between resumings
checkpoint_dir = "./checkpoint/"
checkpoint_name = "last.pth"
checkpoint_path = checkpoint_dir + checkpoint_name

os.makedirs(checkpoint_dir, exist_ok=True)

if not os.path.exists(checkpoint_path):
    start = time.time()
    print("\nDownloading last checkpoint...")
    wget.download(CHECKPOINT_URL, checkpoint_path)
    print("\nDownloaded last checkpoint in {}s".format(round(time.time() - start, 3)))
```
